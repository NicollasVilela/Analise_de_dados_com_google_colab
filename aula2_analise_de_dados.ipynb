```python
import pandas as pd
import numpy as np

df = pd.read_csv('/content/enem_simulado_brasil.csv')
df.head()
```

```python
analise_escola = df.groupby('tipo_escola')[['nota_matematica','nota_linguagens','nota_ciencias','nota_humanas','nota_redacao','media_final']].mean()
display(analise_escola)
```

```python
media_estado = df.groupby('estado')['media_final'].mean().sort_values(ascending=False)
display(media_estado.head())

top10 = df.sort_values('media_final', ascending=False).head(10)
display(top10)
```
